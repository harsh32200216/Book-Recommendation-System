```python
num_rating = rwn.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns = {'Book-Rating': 'num-ratings'}, inplace = True)
num_rating.head()
```

```python
avg_rating = rwn.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns = {'Book-Rating': 'avg-ratings'}, inplace = True)
avg_rating.head()
```

```python
popular = num_rating.merge(avg_rating, on = 'Book-Title')
popular = popular[popular['num-ratings'] >= 250].sort_values('avg-ratings', ascending = False).head(50)
popular
```

```python
filtered_ratings = rwn[rwn['User-ID'].isin(x.keys())]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']
```

```python
y = y[y >= 50]
y
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(y.keys())]
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
pt.fillna(0, inplace = True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_score = cosine_similarity(pt)
```

```python
from difflib import SequenceMatcher

def recommend(book_name):
    ratio = 0
    index = -1
    for x,i in enumerate(pt.index):
        if(SequenceMatcher(None, book_name, i).ratio() > ratio):
            ratio = SequenceMatcher(None, book_name, i).ratio()
            index = x
    similar_items = sorted(list(enumerate(sim_score[index])), key = lambda x: x[1], reverse = True)[:10]
    data = []
    for i in similar_items:
        items = []
        temp = books[books['Book-Title'] == pt.index[i[0]]]
        temp = temp.drop_duplicates('Book-Title')
        items.append(temp['Book-Title'].values[0])
        items.append(temp['Book-Author'].values[0])
        items.append(temp['Image-URL-M'].values[0])
        data.append(items)
    return data
```

```python
recommend('Message in a Bottle')
```

```python
recommend('Harry Potter')
```

```python
recommend('Animal Farm')
```

```python
recommend('The Da Vinci Code')
```
